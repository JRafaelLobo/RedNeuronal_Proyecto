```python
import numpy as np

# Cargar el dataset de prueba
data_test = np.load("mnist_test.npz")
images_test = data_test["images"]
labels_test = data_test["labels"]

# Mostrar la forma de los datos
print("Shape imágenes test:", images_test.shape)
print("Shape etiquetas test:", labels_test.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))

for i in range(15):
    plt.subplot(3, 5, i+1)
    plt.imshow(images_test[i], cmap="gray")
    plt.title(f"Label: {labels_test[i]}")
    plt.axis("off")

plt.tight_layout()
plt.show()

```
